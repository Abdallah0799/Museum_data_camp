```python
import pandas as pd
```

```python
df = pd.read_csv('data/museum_data.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)

df.head(10)
```

```python
pd.DataFrame(df.groupby("Historical Period").size())
```

```python
columns = ['Historical Period'] #, 'Object Begin Date']

df_grouped_by = df.groupby(columns)

df_balanced = df_grouped_by.apply(lambda x: x.sample(df_grouped_by.size().min()).reset_index(drop=True))
df_balanced = df_balanced.droplevel(columns)

df_balanced
```

```python
pd.DataFrame(df_balanced.groupby("Historical Period").size())
```

```python
from sklearn.model_selection import train_test_split

periods = df_balanced["Historical Period"].unique()
df_train, df_test= train_test_split(df_balanced[df_balanced["Historical Period"] == periods[0]])
for period in periods[1:]:
    df_train_p, df_test_p = train_test_split(df_balanced[df_balanced["Historical Period"] == period])
    df_train = pd.concat([df_train, df_train_p])
    df_test = pd.concat([df_test, df_test_p])
```

```python
pd.DataFrame(df_train.groupby("Historical Period").size())
```

```python
pd.DataFrame(df_test.groupby("Historical Period").size())
```

```python
df_train.to_csv('train.csv')
df_test.to_csv('test.csv')
```
